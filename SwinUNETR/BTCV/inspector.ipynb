```python
import json
import os
import re

def strip_extension(filename):
    # Remove both .nii and .gz extensions
    base = os.path.basename(filename)
    if base.endswith(".nii.gz"):
        return base.replace(".nii.gz", "")
    return os.path.splitext(base)[0]

def filter_by_id_limit(id_list, max_id=500):
    filtered = []
    for item in id_list:
        match = re.search(r'AMOS_(\d+)', item)
        if match and int(match.group(1)) <= max_id:
            filtered.append(item)
    return filtered

# Load the original JSON
with open("/home/jovianto/amos22/dataset.json", "r") as f:
    data = json.load(f)

# Extract and filter IDs
train_ids = [strip_extension(item["image"]) for item in data.get("training", [])]
val_ids = [strip_extension(item["image"]) for item in data.get("validation", [])]

train_ids = filter_by_id_limit(train_ids, 500)
val_ids = filter_by_id_limit(val_ids, 500)

# Format output
output_data = [{"train": train_ids, "val": val_ids}]

# Save to new JSON
with open("/home/jovianto/nnUNet_preprocessed/Dataset218_AMOS2022_postChallenge_task1/split.json", "w") as f:
    json.dump(output_data, f, indent=2)
```
